As there are 1474 null values in the rating column so we have to replace these null values with their median

```python
def replace_median(column):
  return column.fillna(column.median)
```

```python
df.Rating = df ['Rating'].transform(replace_median)
```

```python
df.isnull().sum()
```

Data Cleaning

```python
df.isnull().sum()
```

```python
print(df['Type'].mode())
print(df['Current Ver'].mode())
print(df['Android Ver'].mode())
```

```python
df['Type'].fillna(str(df['Type'].mode().values[0]),inplace=True)
df['Current Ver'].fillna(str(df['Current Ver'].mode().values[0]),inplace=True)
df['Android Ver'].fillna(str(df['Android Ver'].mode().values[0]),inplace=True)
df['Content Rating'].fillna(str(df['Content Rating'].mode().values[0]),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Price']=df['Price'].apply((lambda x: str(x).replace('$','') if '$' in str(x) else str(x)))#to make the data frame look nice
df['Reviews']=pd.to_numeric(df['Reviews'],errors='coerce')#to convert all the reviews into numerical value 
df['Price']=pd.to_numeric(df['Price'],errors='coerce')#to convert all the prices into numerical value
```

```python
df['Installs']=df['Installs'].apply(lambda x: str(x).replace('+','')if '+' in str(x) else str(x))#to make the data frame look nice
df['Installs']=df['Installs'].apply(lambda x: str(x).replace(',','')if ',' in str(x) else str(x))#to make the data frame look nice
df['Installs']=pd.to_numeric(df['Installs'],errors='coerce')#to convert the Install column into numerical value
```

Data Visualization

```python
 grp= df.groupby('Category')
 x=grp['Installs'].agg(np.mean)
 y=grp['Price'].agg(np.sum)
 z=grp['Reviews'].agg(np.mean)
 print(x)
 print(y)
 print(z)
```

```python
plt.figure(figsize=(20,10))
plt.plot(x,'ro',color='black')
plt.xticks(rotation=90)
plt.title('Category vs Installs')
plt.xlabel('Categories------------>')
plt.ylabel('Installs------------->')
plt.show
```

```python
plt.figure(figsize=(16,5))
plt.plot(x,'r--',color='indigo')
plt.xticks(rotation=90)
plt.title('Category vs Pricing')
plt.xlabel('Categories------------>')
plt.ylabel('Reviews------------>')
plt.show
```

```python
plt.figure(figsize=(16,5))
plt.plot(z,'g^',color='green')
plt.xticks(rotation=90)
plt.title('Category vs Reviews')
plt.xlabel('Categories------------>')
plt.ylabel('Prices------------>')
plt.show
```

```python
# ploting countplot for "Category" column
```

```python
plt.figure(figsize=(14,7))
plt.style.use("fivethirtyeight")
snb.countplot('Category',data=df,
              order=df.Category.value_counts().head(10).index)
plt.xlabel("Category-------------->", fontsize=15)
plt.ylabel("Genres-------------->", fontsize=15)
plt.title("Count of applications in each category ")
plt.xticks(rotation=90)
plt.show();
```

```python
plt.figure(figsize=(10,7))
plt.style.use("fivethirtyeight")
snb.countplot("Content Rating",data=df)
plt.xlabel("Content Rating---------------->",fontsize=15)
plt.ylabel("No. of Apps---------------->",fontsize=15)
plt.title("Comparison of content rating(Number of Apps in Market)")
plt.show();
```
